# how *e.Rmag* depends on *ApDRmag* and *mumax*?

```python
fig, axes = plt.subplots(1, 2, constrained_layout=True)
for ax, x_column in zip(axes, ['ApDRmag', 'mumax']):
    sns.scatterplot(x=x_column, y='e.Rmag', data=data, ax=ax)
```

```python
sns.scatterplot(data['mumax'], np.log(data['e.Rmag']))
plt.title('mumax vs log(e.Rmag)')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lreg = LinearRegression()
lreg.fit(data['mumax'].values[:, None], data['Rmag'])
r2_score(data['Rmag'], lreg.predict(data['mumax'].values[:, None]))
```

```python
lreg = LinearRegression()
X = data[['mumax', 'ApDRmag']]
lreg.fit(X, data['Rmag'])
r2_score(data['Rmag'], lreg.predict(X))
```

```python
y = data['Rmag']
y_pred = lreg.predict(X)
resid = y - y_pred
ax = sns.scatterplot(y_pred, (resid - resid.mean()) / (resid.max() - resid.min()))
ax.fill_between(y_pred, -data['e.Rmag'], data['e.Rmag'], alpha=0.5)
```
